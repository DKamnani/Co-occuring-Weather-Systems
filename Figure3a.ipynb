```python
co_masked = co_masked.where((co_masked >= 1) & (co_masked <= 15))
```

```python
ds = xr.Dataset({
    'co': co_masked,
    'ar_cat': AR_event_category
})
```

```python
df = ds.to_dataframe().reset_index()
```

```python
df = df.dropna(subset=['co'])
```

```python
#Remapping AR cat 6 and 7 to be included in 5
df['ar_cat'] = df['ar_cat'].replace({6: 5, 7: 5})
```

```python
#Annual counts
counts = df.groupby(['latitude', 'longitude', 'co']).size().reset_index(name='count')
```

```python
import pandas as pd
def get_mode(group):
    max_count = group['count'].max()
    winners = group[group['count'] == max_count]
    
    return pd.Series({
        'dominant_co': winners['co'].iloc[0],
        'tie': len(winners) > 1
    })

dominant_annual = counts.groupby(['latitude', 'longitude']).apply(get_mode).reset_index()
```

```python
#Per cat
counts_ar = df.groupby(['latitude', 'longitude', 'ar_cat', 'co']).size().reset_index(name='count')
```

```python
dominant_ar = counts_ar.groupby(['latitude', 'longitude', 'ar_cat']).apply(get_mode).reset_index()
```

```python
merged = dominant_ar.merge(
    dominant_annual,
    on=['latitude', 'longitude'],
    suffixes=('_ar', '_annual')
)

diff = merged[merged['dominant_co_ar'] != merged['dominant_co_annual']]
```

```python
dominant_annual_xr = dominant_annual.set_index(['latitude','longitude'])['dominant_co'].to_xarray()
```

```python
#plotting
# Count occurrences per pixel + category
counts_all = (
    df.groupby(['latitude','longitude','co'])
    .size()
    .reset_index(name='count')
)

# Total counts per pixel
total_all = counts_all.groupby(['latitude','longitude'])['count'].sum()

# Max counts per pixel
max_all = counts_all.groupby(['latitude','longitude'])['count'].max()

# Percentage
dominant_percentage_all = (max_all / total_all * 100).reset_index(name='percentage')

# Convert to xarray
dominant_percentage_all = dominant_percentage_all.set_index(['latitude','longitude'])['percentage'].to_xarray()
```

```python
dominant_categories = {}

for cat in dominant_ar['ar_cat'].unique():

    subset = dominant_ar[dominant_ar['ar_cat'] == cat]

    xr_map = (
        subset
        .set_index(['latitude', 'longitude'])['dominant_co']
        .to_xarray()
    )

    dominant_categories[cat] = xr_map
```

```python
dominant_percentages = {}

for cat in dominant_ar['ar_cat'].unique():
    
    df_cat = df[df['ar_cat'] == cat]
    
    counts = (
        df_cat.groupby(['latitude','longitude','co'])
        .size()
        .reset_index(name='count')
    )

    # Pivot → full grid per pixel
    pivot = counts.pivot_table(
        index=['latitude','longitude'],
        columns='co',
        values='count',
        fill_value=0
    )

    total = pivot.sum(axis=1)
    maxc = pivot.max(axis=1)

    perc = (maxc / total * 100)

    perc_xr = perc.to_xarray()

    dominant_percentages[cat] = perc_xr
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np
import matplotlib.colors as mcolors
from matplotlib.patches import Patch

# -----------------------------
# CATEGORY LABELS
# -----------------------------
label_index = {
    1: 'AR', 2: 'Front', 3: 'MCS', 4: 'LPS',
    5: 'AR + Front', 6: 'AR + MCS', 7: 'AR + LPS',
    8: 'Front + MCS', 9: 'Front + LPS', 10: 'MCS + LPS',
    11: 'AR + Front + MCS', 12: 'AR + Front + LPS',
    13: 'AR + MCS + LPS', 14: 'Front + MCS + LPS', 15: 'All'
}
# Panel labels
panel_labels = ['1(a)', '1(b)', '1(c)', '1(d)', '1(e)', '1(f)']

#Lat and lon
def add_latlon_labels(ax, show_left=True, show_bottom=True):
    gl = ax.gridlines(
        crs=ccrs.PlateCarree(),
        draw_labels=True,
        linewidth=0.0,     # 🔥 hides grid lines
        color='none',      # 🔥 ensures no lines visible
    )

    # Control where labels appear
    gl.top_labels = False
    gl.right_labels = False
    gl.left_labels = show_left
    gl.bottom_labels = show_bottom

    # Style
    gl.xlabel_style = {'size': 12}
    gl.ylabel_style = {'size': 12}


# -----------------------------
# SELECT PANELS
# -----------------------------
categories_to_plot = [1, 2, 3, 4, 5]

# -----------------------------
# ANNUAL DATA
# -----------------------------
all_category = dominant_annual_xr

lon_min = float(all_category.longitude.min())
lon_max = float(all_category.longitude.max())
lat_min = float(all_category.latitude.min())
lat_max = float(all_category.latitude.max())

extent = [lon_min, lon_max, lat_min, lat_max]
# -----------------------------
# ORIGINAL COLOR SCHEME (DYNAMIC)
# -----------------------------
unique_vals = np.unique(all_category.values[~np.isnan(all_category.values)])
existing_labels = {i: label_index[i] for i in unique_vals if i in label_index}

cmap = plt.get_cmap("tab20", len(existing_labels))
bounds = list(existing_labels.keys()) + [max(existing_labels.keys()) + 1]
norm = mcolors.BoundaryNorm(bounds, cmap.N)

# -----------------------------
# FIGURE SETUP (2x3)
# -----------------------------
fig, axs = plt.subplots(
    nrows=2,
    ncols=3,
    figsize=(18, 6.5),
    subplot_kw={'projection': ccrs.PlateCarree()}
)

axs = axs.ravel()

# -----------------------------
# HELPER FUNCTION
# -----------------------------
def make_rgba(data_xr, alpha_xr):
    Z = data_xr.values
    A = alpha_xr.values

    # Fix upside-down issue if needed
    if data_xr.latitude.values[0] > data_xr.latitude.values[-1]:
        Z = np.flipud(Z)
        A = np.flipud(A)

    rgba = cmap(norm(Z))
    rgba[..., -1] = A

    return rgba

# -----------------------------
# PANEL 1: ANNUAL
# -----------------------------
ax = axs[0]

alpha_all = (dominant_percentage_all / 100.0).clip(min=0.1, max=1.0).fillna(0.0)

rgba = make_rgba(all_category, alpha_all)

ax.imshow(
    rgba,
    origin="lower",
    extent= extent,
    transform=ccrs.PlateCarree(),
    #aspect="auto"
)
ax.set_aspect('equal', adjustable='box')
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linewidth=0.5)
ax.set_title("Annual", fontsize=18)
ax.text(
    0.02, 0.95,
    panel_labels[0],
    transform=ax.transAxes,
    fontsize=14,
    va='top',
    ha='left'
)
add_latlon_labels(ax, show_left=True, show_bottom=True)
# -----------------------------
# PANELS 2–6: AR CATEGORY 1–5
# -----------------------------
for i, cat in enumerate(categories_to_plot, start=1):

    ax = axs[i]

    dom = dominant_categories[cat]
    perc = dominant_percentages[cat]

    alpha = (perc / 100.0).clip(min=0.05, max=1.0).fillna(0.0)

    rgba = make_rgba(dom, alpha)

    ax.imshow(
        rgba,
        origin="lower",
        extent= extent,
        transform=ccrs.PlateCarree(),
        aspect="auto"
    )
    ax.set_aspect('equal', adjustable='box')
    ax.coastlines()
    ax.add_feature(cfeature.BORDERS, linewidth=0.5)
    ax.set_title(f'AR Scale {cat}', fontsize=18)
    ax.text(
    0.02, 0.95,
    panel_labels[i],
    transform=ax.transAxes,
    fontsize=14,
    va='top',
    ha='left'
)
    add_latlon_labels(
    ax,
    show_left=(i % 3 == 0),   # left column only
    show_bottom= True     # bottom row only
)
# -----------------------------
# LEGEND (ONLY SHOWN CATEGORIES)
# -----------------------------
legend_elements = []

for cat, lbl in existing_labels.items():
    color = cmap(norm(cat))
    legend_elements.append(
        Patch(facecolor=color, edgecolor='k', label=lbl)
    )
"""
fig.legend(
    handles=legend_elements,
    title="Dominant Pattern",
    loc="center right",
    bbox_to_anchor=(1.10, 0.2),
    fontsize=10
)
"""
fig.legend(
    handles=legend_elements,
    title="Dominant Pattern",
    loc="center right",
    bbox_to_anchor=(1.12, 0.65),

    fontsize=11,              # label text
    title_fontsize=13,        # title

    frameon=True,
    borderpad=1.2,            # 🔥 padding inside box (bigger box)
    labelspacing=1.0,         # space between entries
    handlelength=1.8,         # width of color patches
    handleheight=1.2,         # height of color patches
    handletextpad=0.8         # space between patch and text
)


from matplotlib.patches import Polygon, Circle

# --------------------------------------------------
# ALPHA PINWHEEL LEGEND (frequency reference)
# --------------------------------------------------

ax_pw = fig.add_axes([0.99, 0.04, 0.12, 0.35])
ax_pw.set_aspect("equal")
ax_pw.axis("off")

ncat = len(existing_labels)

# Smooth circle resolution
ntheta = 300
theta = np.linspace(0, 2*np.pi, ntheta)

# Radial resolution
nr = 120
radii = np.linspace(0.05, 1.0, nr)

# Divide circle into category sectors
theta_edges = np.linspace(0, 2*np.pi, ncat + 1)

levels = [
    (0.26, 0.34, 0.35),   # 30%
    (0.56, 0.64, 0.65),   # 60%
    (0.92, 1.00, 1.00)    # 100%
]
for i, cat in enumerate(existing_labels.keys()):
    base_color = cmap(norm(cat))[:3]

    mask = (theta >= theta_edges[i]) & (theta <= theta_edges[i+1])
    th = theta[mask]

    for r0, r1, alpha_val in levels:
        verts = np.vstack([
            np.column_stack([r0*np.cos(th), r0*np.sin(th)]),
            np.column_stack([r1*np.cos(th[::-1]), r1*np.sin(th[::-1])])
        ])
        poly = Polygon(
            verts,
            closed=True,
            facecolor=(*base_color, alpha_val),
            edgecolor="none"
        )
        ax_pw.add_patch(poly)

# Reference rings + labels
# Reference labels only
for r, lbl in zip([0.3, 0.6, 1.0], ["30%", "60%", "100%"]):
    ax_pw.text(
        0, r,
        lbl,
        ha="center",
        va="bottom",
        fontsize=10,
        fontweight="bold",
        color="black"
    )

ax_pw.set_xlim(-1.1, 1.1)
ax_pw.set_ylim(-1.1, 1.1)



# -----------------------------
# FINALIZE
# -----------------------------
fig.suptitle(
    "Most Dominant Co-occurrence per Grid Cell",
    fontsize=20,
    y=0.98
)
plt.tight_layout()
plt.savefig("Figure3_corrected.pdf", dpi=300, bbox_inches="tight")
plt.show()
```
